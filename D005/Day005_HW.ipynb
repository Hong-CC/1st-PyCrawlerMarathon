```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
```

```python
# æ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½
for idx, data in enumerate(datas):
    print('# ç¬¬ %d ç­†è³‡æ–™åŒ…å«çš„æ¬„ä½æœ‰:%s' %(idx + 1, data.keys()))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for idx, data in enumerate(datas):
    print('ç¬¬ %d ç­†è³‡æ–™ï¼Œæ¨™é¡Œï¼š%sã€è²¼æ–‡æ™‚é–“ï¼š%sã€ç•™è¨€äººæ•¸ï¼š%sã€æŒ‰è®šäººæ•¸ï¼š%s' 
          %(idx +1, data['title'], data['createdAt'], data['commentCount'], data['likeCount']))
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

# ç†±é–€
commentCount = 0
likeCount = 0
for data in datas:
    commentCount += data['commentCount']
    likeCount += data['likeCount']
avg_commentCount = round(commentCount/len(datas),2)
avg_likeCount = round(likeCount/len(datas),2)
print('ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ï¼š%s å¹³å‡æŒ‰è®šäººæ•¸ï¼š%s' %(avg_commentCount, avg_likeCount))
```
