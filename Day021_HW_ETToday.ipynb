## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd

url = 'https://www.ettoday.net/news/news-list.htm'
```

```python
today = time.strftime('%Y/%m/%d')
today
```

```python
browser = webdriver.Chrome(executable_path = 'chromedriver')
browser.get(url)
```

```python
todayNews = []
keepScroll = True

while keepScroll:
    browser.execute_script('window.scrollBy(0,10000)')
    time.sleep(3)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source)
    for entry in soup.find(class_ = 'part_list_2').find_all('h3'):
        if entry.find(class_ = 'date').text[:10] != today:
            keepScroll = False
        else:
            print(entry.find(class_ = 'date').text,
                 entry.find('a').text)
            todayNews.append({
                'time' : entry.find(class_ = 'date').text,
                'tag' : entry.find('em').text,
                'title' : entry.find('a').text
            })
    
```

```python
df = pd.DataFrame.from_dict(todayNews)
df
```

## 2. 取出現在時間兩小時內的新聞

```python
import numpy as np
condition = np.logical_and(df['time'] > '2019/12/28 14:00', df['time'] <= '2019/12/28 16:00')
df[condition]
```

## 3. 根據範例，取出三天前下午三點到五點的新聞  

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time
from datetime import timedelta, datetime
import pandas as pd
import numpy as np

url = 'https://www.ettoday.net/news/news-list.htm'
```

```python
dby = datetime.today() + timedelta(-2)
dby = dby.strftime('%Y/%m/%d')
dby_month = dby[5:7]
dby_day = dby[8:]
```

```python
browser = webdriver.Chrome(executable_path = 'chromedriver')
browser.get(url)
```

```python
# 先將頁面切到12/26

selectMonth = Select(browser.find_element_by_id('selM'))
selectMonth.select_by_value(dby_month)

selectDay = Select(browser.find_element_by_id('selD'))
selectDay.select_by_value(dby_day)
```

```python
browser.find_element_by_id('button').click()
```

```python
# 把整天的都抓出來，再從中抓出特定時間的文章

dbyNews = []
keepScroll = True

while keepScroll:
    browser.execute_script('window.scrollBy(0, 10000)')
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source)
    for entry in soup.find(class_ = 'part_list_2').find_all('h3'):
        if entry.find(class_ = 'date').text[:10] != dby:
            keepScroll = False
        else:
            dbyNews.append({
                'time': entry.find(class_ = 'date').text,
                'tag': entry.find('em').text,
                'title': entry.find('a').text
            })
```

```python
df = pd.DataFrame.from_dict(dbyNews)
df
```

```python
condition = np.logical_and(df['time'] > '2019/12/26 15:00', df['time'] < '2019/12/26 17:00')
df[condition]
```

```python
condition = []
for i in range(df.shape[0]):
    condition.append('博恩夜夜秀' in df['title'][i])
df[condition]
```
